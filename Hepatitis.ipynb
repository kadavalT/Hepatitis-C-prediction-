```python
#Data Cleaning
#Deleting rows which have more than one null value
newdataset=dataset.drop(dataset.index[[584,590,603]])
print(newdataset)
```

```python
#Data replacing
#Final data
Finaldata=newdataset.interpolate()
print(Finaldata)
```

```python
#Preprocessing
#Assinging the values
Finaldata['Category'].loc[Finaldata['Category'].isin(["1=Hepatitis","2=Fibrosis", "3=Cirrhosis"])] = 1
Finaldata['Category'].loc[Finaldata['Category'].isin(["0=Blood Donor", "0s=suspect Blood Donor"])] = 0

Finaldata['Sex'].loc[Finaldata['Sex']=='m']=1
Finaldata['Sex'].loc[Finaldata['Sex']=='f']=0

#Converting the data type
Finaldata['Category']=Finaldata['Category'].astype('int')
Finaldata['Sex']=Finaldata['Sex'].astype('int')
Finaldata
```

```python
Finaldata.info()
```

```python
#kNN Model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
np.random.seed(45)
X= Finaldata.drop('Category', axis=1)
y= Finaldata['Category']
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.30)
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
y_predict = knn.predict(x_test)
accuracy_score(y_test,y_predict)
```

```python
from sklearn import metrics

conf_matrix=metrics.confusion_matrix(y_test,y_predict)
print(conf_matrix)
```
